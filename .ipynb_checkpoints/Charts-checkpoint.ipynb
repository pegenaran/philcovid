```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
plt.close('all')

filename = "./covid-19-data/public/data/ecdc/full_data.csv"
df = pd.read_csv(filename)
df.head()
# [print(x) for x in df.columns.tolist()]
```

```python
asialist = ['Philippines','Indonesia','Japan','Thailand','Malaysia'] #,'Singapore'
asia_df = df.loc[df['location'] == asialist[3]].copy()
asia_df.set_index('date', inplace=True)
startdate = str(pd.Timestamp('2020-03-01').date())
today = dt.datetime.now()
enddate = str(today.date())
print(startdate, enddate)
asia_df = asia_df.loc[startdate:enddate]
asia_df.tail()

```

```python
%matplotlib inline

startdate = str(pd.Timestamp('2020-03-01').date())
today = dt.datetime.now()
enddate = str(today.date())

newplot_df = pd.DataFrame()
ctd_cols_list = []
for c in asialist:
    asia_df = df.loc[df['location'] == c].copy()
    asia_df.set_index('date', inplace=True)
    asia_df = asia_df.loc[startdate:enddate]    
    asia_df['cases_shift'] = asia_df['total_cases'].shift(21)
    # CASES TO DEATH RATIO
    ctd_header = 'CTD_' + c
    asia_df[ctd_header] = round(asia_df['cases_shift'] / asia_df['total_deaths'],2)
    asia_df.drop(axis=1, labels=['cases_shift'], inplace=True)
    ctd_cols_list.append(ctd_header)
    if newplot_df.empty:
        newplot_df = asia_df[[ctd_header]]
    else:
        newplot_df = newplot_df.merge(asia_df[[ctd_header]], how='inner', left_index=True, right_index=True)

newplot_df.plot()
```
